```python
import re
import numpy as np
import pandas as pd
```

```python
Dims = [100, 200, 400, 600, 800, 1000, 1200, 1400, 1600]
threads = [2, 4, 6, 8, 12, 16, 20, 24, 32, 48, 64]

mulmatrix_seq = np.zeros((1, len(Dims)))
f1 = open("./mulmatrix/mulmatrix_seq3.log", "r")
for line in f1.readlines():
    nums = re.findall(r'\d+', line)
    time = re.findall("\d+\.\d+", line)
    if time != []:
        N = int(nums[0])
        t = int(nums[1])
        time = float(time[0])
        j = Dims.index(N)
        mulmatrix_seq[0, j] += time / 3
np.savetxt("./mulmatrix/mulmatrix_seq3.csv", mulmatrix_seq, delimiter="\t", fmt='%f')


mulmatrix_omp = np.zeros((len(threads), len(Dims)))
f2 = open("./mulmatrix/mulmatrix_omp3.log", "r")
for line in f2.readlines():
    nums = re.findall(r'\d+', line)
    time = re.findall("\d+\.\d+", line)
    if time != []:
        N = int(nums[0])
        t = int(nums[1])
        time = float(time[0])
        i = threads.index(t)
        j = Dims.index(N)
        mulmatrix_omp[i, j] += time / 3 
np.savetxt("./mulmatrix/mulmatrix_omp3.csv", mulmatrix_omp, delimiter="\t", fmt='%f')


mulmatrix_pth = np.zeros((len(threads), len(Dims)))
f3 = open("./mulmatrix/mulmatrix_pth3.log", "r")
for line in f3.readlines():
    nums = re.findall(r'\d+', line)
    time = re.findall("\d+\.\d+", line)
    if time != []:
        N = int(nums[0])
        t = int(nums[1])
        time = float(time[0])
        i = threads.index(t)
        j = Dims.index(N)
        mulmatrix_pth[i, j] += time / 3
np.savetxt("./mulmatrix/mulmatrix_pth3.csv", mulmatrix_pth, delimiter="\t", fmt='%f')
```
